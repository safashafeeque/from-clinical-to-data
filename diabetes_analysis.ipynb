```python
# Load pandas tool and assign it as pd
import pandas as pd

# Read and process data in file (diabetes.csv) and store it in the variable data
data = pd.read_csv("diabetes.csv")

# Show the first five entries from the dataset
data.head()
```

```python
# Give information regarding the no. of rows, columns, the different datatypes, and missing values
data.info()
```

```python
# Give a statistical summary (std, min&max, quartiles, mean, count) of all values
data.describe()
```

```python
# Code to replace invalid zero values with NaN (Not a Number)

import numpy as np

columns = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

# In these columns, replace 0 with NaN so Python treats them as missing values
data[columns] = data[columns].replace(0, np.nan)

# Check how many missing values are now present
data.isnull().sum()
```

```python
# Fill missing values with median
data.fillna(data.median(), inplace=True)

# Check again
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt

# Count no. of patients falling into each range for Glucose values and visualise distribution
plt.hist(data["Glucose"], bins=20) #divide the graph into 20 groups
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.title("Distribution of Glucose Levels")
plt.show()
```

```python
import seaborn as sns
#seaborn library compares distributions

# Compare glucose levels between people with and without diabetes
sns.boxplot(x="Outcome", y="Glucose", data=data)
plt.title("Glucose Levels by Diabetes Outcome")
plt.show()
```

- Outcome 0 - no diabetes

- Outcome 1 - diabetes


```python
# Compare BMI between people with and without diabetes
sns.boxplot(x="Outcome", y="BMI", data=data)
plt.title("BMI Distribution by Diabetes Outcome")
plt.show()
```
